# 0.0_list_stadt_nrw

Here I scrap the cities of NRW from the website https://www.nrw-live.de/staedte-nrw 
Tutorial: https://medium.com/technofunnel/web-scraping-with-python-using-beautifulsoup-76b710e3e92f 

```python
import requests
import pandas as pd
import janitor
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.nrw-live.de/staedte-nrw'

req = requests.get(url)
```

```python
soup = BeautifulSoup(req.content, 'html5lib') 
```

```python
list_stadt_nrw = soup.findAll('a') 
list_stadt_nrw # retornou uma lista para cara <a>
```

```python
len(list_stadt_nrw)
```

```python
list_stadt_nrw[87].text # start in 87
```

```python
list_stadt_nrw[482].text # ends here
```

```python
stadt_nrw = []

for i in range(87, 483):
    t = list_stadt_nrw[i].text
    stadt_nrw.append(t)
```

```python
stadt_nrw
```

```python
type(stadt_nrw)
```

```python
df_stadt_nrw = pd.DataFrame({'stadt':stadt_nrw})
df_stadt_nrw
```

```python
#Exportando em CSV
df_stadt_nrw.to_csv(r'df_stadt_nrw.csv', index = False)
```
